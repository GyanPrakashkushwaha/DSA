```python
class Solution:
    def evalRPN(self, tokens: list[str]) -> int:
        st = []
        for token in tokens:
            print(st)
            if token == "+":
                calculation = st.pop() + st.pop()
                st.append(calculation)
            elif token == "-":
                calculation = st.pop() - st.pop()
                st.append(calculation)
            elif token == "*":
                calculation = st.pop() * st.pop()
                st.append(calculation)
            elif token == '/':
                num1 = st.pop()
                calculation = num1 / st.pop()
                st.append(calculation)
            else:
                st.append(token)
        return st[-1]
```

```python
sol = Solution()
sol.evalRPN(["2","1","+","3","*"])
```
