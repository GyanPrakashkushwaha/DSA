```python
def min_total_distances(N, A):
    prefix = [0] * N
    prefix[0] = A[0]

    # Step 1: Compute prefix sum
    for i in range(1, N):
        prefix[i] = prefix[i-1] + A[i]

    out = []

    k = N - 2  # number of friends to invite

    for i in range(N):
        # Determine the window of k friends around i
        left = max(0, i - k // 2)
        right = left + k

        # If window goes out of bounds, shift it back
        if right >= N:
            right = N - 1
            left = right - k

        # Sum of positions of invited friends
        total_sum = prefix[right]
        if left > 0:
            total_sum -= prefix[left - 1]

        # Distance = sum of |A[j] - A[i]| = total_sum - k*A[i]
        # since A[i] is the meeting point and is subtracted k times
        distance = abs(total_sum - k * A[i])
        out.append(distance)

    return out
min_total_distances(5, l1)
```
