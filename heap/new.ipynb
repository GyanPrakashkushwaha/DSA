```python
class Solution:
    def frequencySort(self, s: str) -> str:
        freq = Counter(s)    
        heap = []
        for ch, f in freq.items():
            heapq.heappush(heap, (-f, ch))
        print(heap)
        res = ''
        while heap:
            f, ch = heapq.heappop(heap)
            res = res +  (-f * ch)

        return res
    
s = "tree"
Solution().frequencySort(s)
```

```python
from typing import List
class Solution:
    def areSimilar(self, mat: List[List[int]], k: int) -> bool:
        n = len(mat[0])
        k = k % n
        mat2 = []
        for i in range(len(mat)):
            row = mat[i]
            if i % 2 == 0:
                lst = row[k:] + row[:k]
                mat2.append(lst)
            else:
                if n % 2 == 0:
                    lst = row[n-k:] + row[:k]
                else:
                    lst = row[n-k+1:] + row[:k]
                mat2.append(lst)
        print(mat2)
        return mat == mat2
Solution().areSimilar([[7,7],[10,10],[4,4]], 2)
```

```python
from typing import List
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        top = 0
        bot = len(matrix) - 1
        mid = 1
        while top < bot:
            mid = top + (bot - top) // 2
            if matrix[mid][0] <= target and matrix[mid][-1] >= target:
                break
            elif matrix[mid][0] > target:
                bot = mid
            else:
                top = mid + 1

        s, e = 0, len(matrix[0]) -1
        row = matrix[mid]
        print(row)
        while s < e:
            m = s + (e - s)//2
            print(row[m])
            if row[m] == target:
                return True
            elif row[m] > target:
                e = m - 1
            else:
                s = m + 1

        return False

Solution().searchMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]], 3)
```

```python
from typing import List
import heapq
class Solution:
    def kSmallestPairs(self, nums1: List[int], nums2: List[int], k: int) -> List[List[int]]:
        heap1, heap2 = [], [] # Min Heaps
        c1, c2 = len(nums1), len(nums2)
        for n in nums1:
            heapq.heappush(heap1, (n, c1))
        for n in nums2:
            heapq.heappush(heap2, (n, c2))

        res = []
        for _ in range(k):
            n1, c1 = heapq.heappop(heap1)
            n2, c2 = heapq.heappop(heap2)
            res.append([n1, n2])
            c1 -= 1
            c2 -= 1
            if c1 != 0: heapq.heappush(heap1, (n1, c1))
            if c2 != 0: heapq.heappush(heap2, (n2, c2))
            print(heap1, heap2)

        return res

Solution().kSmallestPairs(nums1 = [1,1,2], nums2 = [1,2,3], k = 2)
```
