```python
class Solution:
    def twoCitySchedCost(self, costs: list[list[int]]) -> int:
        sortedA = sorted(costs, key=lambda x: x[0])
        sortedB = sorted(costs, key=lambda x: x[1])
        print(sortedA, sortedB)
        total = 0
        print((len(costs)//2))
        for i in range((len(costs)//2)):
            total += sortedA[i][0] + sortedB[i][1]
        
        return total

costs = [[10,20],[30,200],[400,50],[30,20]]
Solution().twoCitySchedCost(costs)
```
