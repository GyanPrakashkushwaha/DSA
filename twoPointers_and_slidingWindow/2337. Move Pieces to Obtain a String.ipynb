```python
class Solution:
    def minimumSteps(self, s: str) -> int:
        balls = list(s)
        i = 0
        swaps = 0
        for j in range(len(s)):
            if balls[j] == "0" and i == j:
                i += 1
            elif balls[j] == "0":
                balls[i], balls[j] = balls[j], balls[i]
                print(balls, j, i)
                swaps += (j - i)
                i += 1
                
        print(balls)
        return swaps
    
Solution().minimumSteps("11000111")
```
