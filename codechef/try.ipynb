```python
def solution(n, x):
    # print(n, x)
    cost = 0
    sales = 0
    for i in range(1, n+1):
        print(i)
        if x <= i:
            sales += i
            # print(sales)
            cost += x
    # print(sales, cost)
    return sales - cost

solution(n = 5, x= 2)
```

```python
def sol(milk, dark):
    sat = 0
    for i in range(milk + dark):
        print('BEFORE--',milk, dark)
        if milk != 0:
            milk -= 1
        else:
            dark -= 1
        
        if milk != dark:
            sat += 1
        print('AFTER--',milk, dark)
        print('SATIFCASION', sat, '\n')
    
    return sat

sol(1, 2)
```
