```python
def pr_binary(num):
    # Loop from bit 10 down to 0
    for i in range(10, -1, -1):
        # Shift the i-th bit to the 0th position and check if it's 1 or 0
        print((num >> i) & 1, end="") # right shift
    print() # Print a newline

pr_binary(13) # Output: 00000001101
```
