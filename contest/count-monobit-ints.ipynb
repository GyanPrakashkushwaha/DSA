```python
class Solution:
    def countMonobit(self, n: int) -> int:
        val = 1
        count = 1
        while val <= n:
            count += 1
            val = val << 1 | 1

        return count

soln = Solution()
soln.countMonobit(2)  
```

```python
grid = [
        [".",".",".",".","5",".",".","1","."],
        [".","4",".","3",".",".",".",".","."],
        [".",".",".",".",".","3",".",".","1"],
        ["8",".",".",".",".",".",".","2","."],
        [".",".","2",".","7",".",".",".","."],
        [".","1","5",".",".",".",".",".","."],
        [".",".",".",".",".","2",".",".","."],
        [".","2",".","9",".",".",".",".","."],
        [".",".","4",".",".",".",".",".","."]
    ]

class Solution:
    def isValidSudoku(self, board: list[list[str]]) -> bool:
        
        def isValidGrid(r, c):
            vals = []
            for i in range(r, r+3):
                for j in range(c, c+3):
                    num =  0 if board[i][j] == '.' else int(board[i][j])
                    vals.append(num)
            
            vals = sorted(vals)
            prev = vals[0]
            for i in range(1, len(vals)):
                next = vals[i]
                if vals[i] and prev == next:
                    return False
            return True
        
        # Step-1 - Check for rows
        for i in range(9):
            vals = [0]*10
            row = board[i]
            for num in row:
                num = -1 if num == '.' else int(num)
                if not num == -1:
                    if vals[num]:
                        return False
                    else:
                        vals[num] = 1
                        
        # step-2 - check for columns
        for i in range(9):
            vals = [0]*10
            col = [row[i] for row in board]
            for num in col:
                num = -1 if num == '.' else int(num)
                if not num == -1:
                    if vals[num]:
                        return False
                    else:
                        vals[num] = 1
                        
        # step-3 check for grids. 
        for i in [0, 3, 6]:
            for j in [0, 3, 6]:
                if not isValidGrid(i, j):
                    return False
        
        return True
    
soln = Solution()
soln.isValidSudoku(grid)
```
