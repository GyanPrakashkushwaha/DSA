```python
class Solution:
    def rotateElements(self, nums: list[int], k: int) -> list[int]:
        # step-1
        pos = {}
        arr = []
        for i in range(len(nums)):
            if nums[i] < 0:
                pos[nums[i]] = i
            else:
                arr.append(nums[i])    
        # step-2
        k = k % len(arr)
        def reverse(l, r):
            while l < r:
                arr[l], arr[r] = arr[r], arr[l]
                l += 1
                r -= 1
        
        reverse(0, k-1)
        reverse(k, len(arr)-1)
        reverse(0, len(arr)-1)

        # step-3
        for ele, idx in pos.items():
           arr.insert(idx, ele)

        return arr
    
soln = Solution()
soln.rotateElements(nums = [1,-2,3,-4], k = 3)
```
